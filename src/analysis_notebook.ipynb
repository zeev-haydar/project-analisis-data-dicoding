## Check Null Values

```python
for key, val in data.get_datas().items():
    print(f"Dataset {key}\n")
    for col in val.columns:
        print(f"Number of null values for column {col} = {val[col].isna().sum()}")
    print("\n====================================\n")
```

## Impute Null Values with mean for numeric, and mode for categorical

```python
import warnings
warnings.filterwarnings('ignore')

for key, val in data.get_datas().items():
    categorical_features = val.select_dtypes(include=['object']).columns
    numerical_features = val.select_dtypes(exclude=['object']).columns
    
    # Impute categorical features with the mode
    for col in categorical_features:
        if val[col].isnull().any():
            mode_value = val[col].mode()[0]
            val[col].fillna(value=mode_value, inplace=True)

    # Impute numerical features with the median
    for col in numerical_features:
        if val[col].isnull().any():
            median_value = val[col].median()
            val[col].fillna(value=median_value, inplace=True) 
    
```

## See products that have the highest orders

### Show product name english translation

```python
product_name = data.get_data("product_category_name")["product_category_name_english"].drop_duplicates()
for i in range(20):
    print(product_name[i])
```

As you can see, the product name is using the underscore as word separator. It is not looking good for displaying information. Therefore, I replace the underscore with whitespace

```python
data.get_data("product_category_name")["product_category_name_english"] = data.get_data("product_category_name")["product_category_name_english"].apply(lambda x: x.replace("_", " "))
product_name = data.get_data("product_category_name")["product_category_name_english"].drop_duplicates()
for i in range(20):
    print(product_name[i])
```

```python
order_items_joined = pd.merge(data.get_data("order_items"), data.get_data("orders"), "inner", on="order_id")
order_item_products = pd.merge(order_items_joined, data.get_data("products"), "inner", "product_id").merge(data.get_data("product_category_name"), "left", "product_category_name")
order_item_products.head()
```

```python
# order_item_products.info()
data.get_data("product_category_name").info()
```

```python
order_item_counts = order_item_products.value_counts(subset="product_category_name_english").reset_index()
order_item_counts
```

```python
top_10_products = order_item_counts.iloc[:10]


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Bar chart on the first subplot (left)
ax1.set_title("Most Bought Product Categories")
ax1.bar(top_10_products["product_category_name_english"], top_10_products["count"], width=0.8)
ax1.set_xticklabels(top_10_products["product_category_name_english"], rotation=45)
ax1.set_ylabel('Count')

# Pie chart on the second subplot (right)
ax2.set_title("Product Category Distribution")
ax2.pie(top_10_products["count"], 
        labels=top_10_products["product_category_name_english"], 
        autopct='%1.1f%%', startangle=90)

plt.tight_layout()
plt.show()
```

Based on the data, "bed bath table" is the most bought product. it accounts for 15.5% of orders

## Customer Segmentation by location

```python
print("Data Customer")
display(data.get_data("customers").head())

print("Data Geolocation")
display(data.get_data("geolocation").head())
```

```python
# Since the prefix code in the geolocation table is not unique, 
# I have decided to group the data by city and average the latitude and longitude of the geolocations in each city.
geolocation_mean_lat_and_long = data.get_data("geolocation").groupby("geolocation_city")[["geolocation_lat", "geolocation_lng"]].agg(["mean"])
geolocation_mean_lat_and_long.columns = geolocation_mean_lat_and_long.columns.get_level_values(0)

cust_segs = pd.merge(data.get_data("customers"), geolocation_mean_lat_and_long.reset_index(), "inner", left_on="customer_city", right_on="geolocation_city")
cust_segs.head(10)
```

### Group By City and show plot for the spread of customers

```python
city_counts = cust_segs['customer_city'].value_counts().reset_index()
city_counts.columns = ['customer_city', 'count']
k = 20
city_counts_top_k = city_counts.iloc[:k]

# Set the figure size
plt.figure(figsize=(10, 8))

# Plot horizontal barplot with Seaborn
plt.barh(city_counts_top_k['customer_city'], city_counts_top_k['count'], color='skyblue')

plt.gca().invert_yaxis()

# Set title and labels
plt.title("Customer Count by City")
plt.xlabel("Count")
plt.ylabel("City")

# Display the plot
plt.show()
```

```python
state_counts = cust_segs['customer_state'].value_counts().reset_index()
state_counts
```

### Display using world map heatmap (Brazil)

```python
import geopandas as gpd
gdf = gpd.read_file("../data/brazil/br_shp/br.shp")
state_counts["customer_state_for_map_merge"] = state_counts['customer_state'].apply(lambda x: "BR"+str(x))
print(state_counts)
```

```python
brazil_map = gdf.merge(state_counts, left_on='id', right_on='customer_state_for_map_merge', how='inner')
# print(brazil_map.isna().sum())

plt.figure(figsize=(16, 16))
ax = brazil_map.plot(column='count', cmap='PuRd', linewidth=1, edgecolor='0.6', legend=True, aspect=1.0)

# Add a title
plt.title('Customer Count by State in Brazil', fontsize=15)
plt.axis('off')

# Annotate state names on the map
for x, y, label, count in zip(brazil_map.geometry.centroid.x, brazil_map.geometry.centroid.y, brazil_map['customer_state_for_map_merge'], brazil_map['count']):
    short_label = label[2:]
    font_color = 'white' if count > brazil_map['count'].quantile(0.97) else 'black'
    ax.annotate(short_label, xy=(x, y), horizontalalignment='center', fontsize=6, color=font_color, weight='bold')

# Show the plot
plt.show()
```

From the Heatmap above, we can conclude that most of the customers live at the southern state
